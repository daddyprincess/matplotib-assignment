## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

```python
Matplotlib is a Python library used for creating static, animated, and interactive visualizations in various formats. It is
one of the most widely used libraries for data visualization in Python.

Matplotlib is used for various purposes, including:

1.Data Exploration: Matplotlib helps visualize data to gain insights and understand patterns.

2.Data Presentation: It is used for creating clear and informative visualizations for reports and presentations.

3.Scientific Research: Scientists and researchers use it to visualize experimental and simulation results.

4.Machine Learning: Matplotlib is often used to visualize the results of machine learning algorithms.

5.Education: It is used in educational settings to teach data visualization concepts.

Five plots that can be created using the Pyplot module of Matplotlib are:

1.Line Plot: Used for visualizing data points connected by straight lines. It is suitable for time series data or data with a
 continuous relationship.

2.Scatter Plot: Shows individual data points as dots, suitable for displaying the relationship between two variables.

3.Bar Chart: Displays categorical data with rectangular bars of varying heights. It is useful for comparing different 
 categories.

4.Histogram: Used for visualizing the distribution of a dataset. It divides data into bins and shows the frequency of values 
 in each bin.

5.Pie Chart: Represents data as a circle divided into slices to display the composition of a whole.

Matplotlib provides many other plot types and extensive customization options for creating a wide range of visualizations.
```

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
A scatter plot is a type of data visualization that displays individual data points as separate dots on a two-dimensional
plane. It is used to show the relationship or correlation between two variables. Each point on the plot represents a single
data observation, with one variable's value determining the horizontal position (x-axis) and the other variable's value
determining the vertical position (y-axis) of the point.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, c='blue', marker='o', label='Data Points')

# Add title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```

## Q3. Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(12, 6))

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, plot 1
plt.plot(x1, y1, label='Line 1')
plt.title('Line Plot 1')
plt.legend()

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, plot 2
plt.plot(x2, y2, label='Line 2')
plt.title('Line Plot 2')
plt.legend()

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, plot 3
plt.plot(x3, y3, label='Line 3')
plt.title('Line Plot 3')
plt.legend()

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, plot 4
plt.plot(x4, y4, label='Line 4')
plt.title('Line Plot 4')
plt.legend()

# Adjust layout and display
plt.tight_layout()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest profit at the top
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 4))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], vert=False, patch_artist=True)
plt.xlabel('Value')
plt.title('Box Plot of Box1 and Box2')
plt.show()
```
